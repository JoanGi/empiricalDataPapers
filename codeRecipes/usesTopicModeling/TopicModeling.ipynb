```python
# Get the data you mau want to use
results = pd.read_excel("rawResults.xlsx")
print(results.head())
```

```python
from bertopic.vectorizers import ClassTfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from transformers import pipeline
from bertopic.representation import TextGeneration


docs = results['uses_data_uses']
# We remove frequent words to create better topic representation
ctfidf_model = ClassTfidfTransformer(reduce_frequent_words=True)
vectorizer_model = CountVectorizer(stop_words="english")

topic_model = BERTopic(vectorizer_model=vectorizer_model, ctfidf_model=ctfidf_model)
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_model.get_topic_info()
```

```python
topics_to_merge = [[-1,0,1,11,13,44]]
topic_model.merge_topics(results['uses_data_uses'], topics_to_merge)
topic_model.get_topic_info()
```

```python
topic_model.reduce_topics(results['uses_data_uses'], nr_topics=20)
topic_model.get_topic_info()
```

```python
hierarchical_topics = topic_model.hierarchical_topics(results['uses_data_uses'])
topic_model.visualize_hierarchy(hierarchical_topics=hierarchical_topics)
```

```python
# Saving the topic model
embedding_model = "sentence-transformers/all-MiniLM-L6-v2"
topic_model.save("./topicModels", serialization="safetensors", save_ctfidf=True, save_embedding_model=embedding_model)
```
